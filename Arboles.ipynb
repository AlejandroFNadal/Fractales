# Árboles

Vamos a construir un arbol, creando una linea vertical, y luego desde su extremo, vamos a crear dos lineas nuevas, en direcciones de 60° y -60°, cuya longitud sea la mitad de la original.


```python
import math
import matplotlib.pyplot as pl
```

```python
def euclidMetric(x1,y1,x2,y2):
    return(math.sqrt((x2-x1)**2 + (y2-y1)**2))
```

```python
depth= 5

class Tree():
    def __init__(self,root):
        self.root = root
        self.children = []
    def addNode(self,obj):
        self.children.append(obj)
class Node:
    def __init__(self,data):
        self.data = data
        self.children = []
    def addNode(self,obj):
        self.children.append(obj)

#trunk = Tree(Node([5,0,5,1])) ##First line
trunk = Node([5,0,5,1])
def addBranches(prev, node, level, depth):
    for i in range(2):
        print(i)
        origx=prev[2]
        origy=prev[3]
        length=euclidMetric(prev[0],prev[1],prev[2],prev[3])/2
        if i == 0:
            angle = 60
            print("right")
        if i ==1:
            angle= 120
            print("left")
        newY= origy+length*math.sin(math.radians(angle))
        newX = origx + length*math.cos(math.radians(angle))
        newNode=Node([origx,origy,newX, newY])
        node.addNode(newNode)
        if level < depth:
            level+=1
            print("Level" +str(level))
            addBranches(newNode.data, newNode,level, depth)
            level-=1
            
addBranches(trunk.data, trunk,0,depth)
```

```python
print(trunk.children[0].data)
print(trunk.children[1].data)
print(trunk.children[0].children[0].data)
print(trunk.children[0].children[1].data)
print(trunk.children[1].children[0].data)
print(trunk.children[1].children[1].data)
print(trunk.children[1].children[1].children[0].data)
print(trunk.children[1].children[1].children[1].data)
print(trunk.children[1].children[0].children[0].children[0].children[0].data)
```

```python
fig = pl.figure(figsize=(12,12))


def goThroughBranches(node):
    pl.plot([node.data[0],node.data[2]],[node.data[1],node.data[3]])
    for child in node.children:
        goThroughBranches(child)

goThroughBranches(trunk)
```
