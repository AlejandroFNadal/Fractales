# Árboles

Vamos a construir un arbol, creando una linea vertical, y luego desde su extremo, vamos a crear dos lineas nuevas, en direcciones de 60° y -60°, cuya longitud sea la mitad de la original.


```python
import math
import matplotlib.pyplot as pl
```

```python
def euclidMetric(x1,y1,x2,y2):
    return(math.sqrt((x2-x1)**2 + (y2-y1)**2))
```

```python
depth= 5

class Tree():
    def __init__(self,root):
        self.root = root
        self.children = []
    def addNode(self,obj):
        self.children.append(obj)
class Node:
    def __init__(self,data):
        self.data = data
        self.children = []
    def addNode(self,obj):
        self.children.append(obj)

#trunk = Tree(Node([5,0,5,1])) ##First line
trunk = Node([5,0,5,1])
def addBranches(prev, node, level, depth):
    for i in range(2):
        print(i)
        origx=prev[2]
        origy=prev[3]
        length=euclidMetric(prev[0],prev[1],prev[2],prev[3])/2
        if i == 0:
            angle = 60
            print("right")
        if i ==1:
            angle= 120
            print("left")
        newY= origy+length*math.sin(math.radians(angle))
        newX = origx + length*math.cos(math.radians(angle))
        newNode=Node([origx,origy,newX, newY])
        node.addNode(newNode)
        if level < depth:
            level+=1
            print("Level" +str(level))
            addBranches(newNode.data, newNode,level, depth)
            level-=1
            
addBranches(trunk.data, trunk,0,depth)
```

```python
print(trunk.children[0].data)
print(trunk.children[1].data)
print(trunk.children[0].children[0].data)
print(trunk.children[0].children[1].data)
print(trunk.children[1].children[0].data)
print(trunk.children[1].children[1].data)
print(trunk.children[1].children[1].children[0].data)
print(trunk.children[1].children[1].children[1].data)
print(trunk.children[1].children[0].children[0].children[0].children[0].data)
```

```python
fig = pl.figure(figsize=(12,12))
fig.suptitle("Arbol con ramas en 60 grados", fontsize=16)

def goThroughBranches(node):
    pl.plot([node.data[0],node.data[2]],[node.data[1],node.data[3]])
    for child in node.children:
        goThroughBranches(child)

goThroughBranches(trunk)
```

### Analisis de la longitud de las ramas

La primera rama tiene una longitud de 1. Las siguientes 2, tienen una longitud de $\frac{1}{2}$ , por lo que sumadas, tienen tambien una longitud de 1. El siguiente nivel tiene ramas de una longitud de un cuarto, y son 4, por lo que tambien tenemos una longitud de 1. 

Se deduce que la longitud de cada iteracion es 1, y por lo tanto, con n iteraciones tendremos una longitud n. Si la figura tiene iteraciones tendiendo a infinito, la longitud sera infinita.

## Arbol con angulos acumulados

Ahora, en lugar de colocar cada rama a 60 grados, llevaremos el angulo de la rama actual a la siguiente iteracion y lo sumaremos

```python
def addBranchesWithAngleSum(prev, node, level, depth, angle):
    for i in range(2):
        origx=prev[2]
        origy=prev[3]
        length=euclidMetric(prev[0],prev[1],prev[2],prev[3])/2
        if i == 0:
            angle = angle - 60
            print("right")
        if i ==1:
            angle= angle + 120
            print("left")
        newY= origy+length*math.sin(math.radians(angle))
        newX = origx + length*math.cos(math.radians(angle))
        newNode=Node([origx,origy,newX, newY])
        node.addNode(newNode)
        if level < depth:
            level+=1
            print("Level" +str(level))
            addBranchesWithAngleSum(newNode.data, newNode,level, depth, angle)
            level-=1
            
```

Prestar atencion a las lineas 8 y 12 donde la suma se realiza.

```python
depth = 12
trunkAngleSum = Node([5,0,5,1])

addBranchesWithAngleSum(trunkAngleSum.data,trunkAngleSum,0,depth,90)
```

```python
fig2 = pl.figure(figsize=(12,12))
fig2.suptitle("Arboles con ramas a 60 grados, con angulos acumulados en cada iteracion}", fontsize=16)

goThroughBranches(trunkAngleSum)
```

Finalmente, hagamos un arbol donde para cada iteracion no se hagan ambas ramas sino que sea aleatorio cual se hará. 

```python
from random import randint
from random import gauss

def addBranchesWithRandomPath(prev, node, level, depth, angle):
    origx=prev[2]
    origy=prev[3]
    length=euclidMetric(prev[0],prev[1],prev[2],prev[3])/2
    i=randint(0,1)
    if i == 0:
        angle = angle - 60
    if i ==1:
        angle= angle + 120
    newY= origy+length*math.sin(math.radians(angle))
    newX = origx + length*math.cos(math.radians(angle))
    newNode=Node([origx,origy,newX, newY])
    node.addNode(newNode)
    if level < depth:
        level+=1
        addBranchesWithRandomPath(newNode.data, newNode,level, depth, angle)
        level-=1
            
```

```python
depth = 5
trunkAngleRandom = Node([5,0,5,1])
trunkAngleRandom.children=[Node([5,1,4.5,1.5]), Node([5,1,5.5,1.5])]
addBranchesWithRandomPath(trunkAngleRandom.children[0].data,trunkAngleRandom.children[0],0,depth,45)
addBranchesWithRandomPath(trunkAngleRandom.children[1].data,trunkAngleRandom.children[1],0,depth,45)
```

```python
fig2 = pl.figure(figsize=(12,12))
fig2.suptitle("Arboles con ramas a 60 grados, con una rama aleatoria en su direccion, partiendo de dos ramas", fontsize=16)

goThroughBranches(trunkAngleRandom)
```

Podemos, ademas, intentar hacer que ambas ramas se toquen, jugando un poco con las distribuciones de probabilidades. Vamos a añadir un parametro tendencia, que nos va a dirigir a la derecha o a la izquierda

```python
def addBranchesWithTendencyPath(prev, node, level, depth, angle, tendency):
    origx=prev[2]
    origy=prev[3]
    length=euclidMetric(prev[0],prev[1],prev[2],prev[3])/1.9
    i=round(gauss(tendency, 0.25))
    if i == 0:
        angle = angle - 35
    if i ==1:
        angle= angle + 70
    newY= origy+length*math.sin(math.radians(angle))
    newX = origx + length*math.cos(math.radians(angle))
    newNode=Node([origx,origy,newX, newY])
    node.addNode(newNode)
    if level < depth:
        level+=1
        addBranchesWithTendencyPath(newNode.data, newNode,level, depth, angle, tendency)
        level-=1
            
```

```python
depth = 9
trunkAngleRandom = Node([5,0,5,1])
trunkAngleRandom.children=[Node([5,1,4.5,1.5]), Node([5,1,5.5,1.5])]
addBranchesWithTendencyPath(trunkAngleRandom.children[0].data,trunkAngleRandom.children[0],0,depth,90,0.25)
addBranchesWithTendencyPath(trunkAngleRandom.children[1].data,trunkAngleRandom.children[1],0,depth,90,0.75)
```

```python
fig2 = pl.figure(figsize=(12,12))
fig2.suptitle("Arboles con ramas a 60 grados, con una rama con tendencia a unirse a la otra", fontsize=16)

goThroughBranches(trunkAngleRandom)
```
