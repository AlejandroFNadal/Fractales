# Curva de Koch

La curva de Koch es un fractal que se forma de la siguiente manera. Primero, dibujamos una linea de longitud 1. Luego, separamos en tres partes, y al segmento del medio, lo dividimos al medio una vez mas, levantando ambos extremos hasta formar un triangulo equilatero. Obtenemos una poligonal de 4 lados. Luego, repetimos este proceso.



Vamos a comenzar, importando librerias

```python
import math
import matplotlib.pyplot as pl

```

Creemos ahora una funcion que cree el fractal. Para asegurarnos de que el triangulo se expande en la direccion correcta, crearemos un centro de la figura desde el cual el fractal siempre se expandira en la direccion contraria

```python
import numpy as np
pi = np.pi
import math
```

```python
def dist_lin(linea):
    x_lin=linea[0]
    y_lin=linea[1]
    dist = math.sqrt(math.pow((x_lin[1]-x_lin[0]),2) + math.pow(y_lin[1]-y_lin[0],2))
    return dist

def dist_lin_x(linea):
    return linea[2][1]-linea[2][0]
def dist_lin_y(linea):
    return linea[3][1]-linea[3][0]

class Linea:
    def __init__(self, f_angle, s_angle, x_s,y_s):
        self.f_angle = f_angle
        self.s_angle = s_angle
        self.x_s = x_s # x points
        self.y_s = y_s # y points
    def dist_x(self):
        return (self.x_s[1] - self.x_s[0])/3
    def dist_y(self):
        return (self.y_s[1] - self.y_s[0])/3
    def dist(self):
        return (math.sqrt(math.pow(self.dist_x()*3,2) + math.pow(self.dist_y()*3,2)))/3
    def dividir(self):
        if self.f_angle == 2*pi/3 and self.s_angle == 0:
            prim_linea = Linea(self.f_angle, self.s_angle, [self.x_s[0], self.x_s[0]+self.dist_x()],[self.y_s[0], self.y_s[0]+self.dist_y()])
            sec_linea = Linea(self.f_angle + pi/3, self.s_angle + pi/3, [prim_linea.x_s[1], prim_linea.x_s[1]+self.dist()*math.cos(self.f_angle)],[prim_linea.y_s[1], prim_linea.y_s[1]+self.dist()*math.sin(self.f_angle)])
            ter_linea = Linea(self.s_angle + pi/3, self.s_angle - pi/3,[sec_linea.x_s[1],sec_linea.x_s[1] + self.dist()*math.cos(self.s_angle)],[sec_linea.y_s[1], sec_linea.y_s[1]+ self.dist()*math.sin(self.s_angle)])
            cuar_linea = Linea(self.f_angle, self.s_angle, [ter_linea.x_s[1], self.x_s[1]], [ter_linea.y_s[1],self.y_s[1]])
            return [prim_linea,sec_linea,ter_linea,cuar_linea]
        if self.f_angle == 0 and self.s_angle == 4*pi/3:
            prim_linea = Linea(self.f_angle, self.s_angle, [self.x_s[0], self.x_s[0]+self.dist_x()],[self.y_s[0], self.y_s[0]+self.dist_y()])
            sec_linea = Linea(self.f_angle + pi/3, self.f_angle - pi/3, [prim_linea.x_s[1], prim_linea.x_s[1]+self.dist()*math.cos(self.f_angle)],[prim_linea.y_s[1], prim_linea.y_s[1]+self.dist()*math.sin(self.f_angle)])
            ter_linea = Linea(self.s_angle + pi/3, self.s_angle - pi/3,[sec_linea.x_s[1],sec_linea.x_s[1] + self.dist()*math.cos(self.s_angle)],[sec_linea.y_s[1], sec_linea.y_s[1]+ self.dist()*math.sin(self.s_angle)])
            cuar_linea = Linea(self.f_angle, self.s_angle, [ter_linea.x_s[1], self.x_s[1]], [ter_linea.y_s[1],self.y_s[1]])
            return [prim_linea,sec_linea,ter_linea,cuar_linea]
        if self.f_angle == 5*pi/3 and self.s_angle == pi/3:
            prim_linea = Linea(self.f_angle, self.s_angle, [self.x_s[0], self.x_s[0]+self.dist_x()],[self.y_s[0], self.y_s[0]+self.dist_y()])
            sec_linea = Linea(self.f_angle - pi/3, self.f_angle + pi/3, [prim_linea.x_s[1], prim_linea.x_s[1]+self.dist()*math.cos(self.f_angle)],[prim_linea.y_s[1], prim_linea.y_s[1]+self.dist()*math.sin(self.f_angle)])
            ter_linea = Linea(self.s_angle - pi/3, self.s_angle + pi/3,[sec_linea.x_s[1],sec_linea.x_s[1] + self.dist()*math.cos(self.s_angle)],[sec_linea.y_s[1], sec_linea.y_s[1]+ self.dist()*math.sin(self.s_angle)])
            cuar_linea = Linea(self.f_angle, self.s_angle, [ter_linea.x_s[1], self.x_s[1]], [ter_linea.y_s[1],self.y_s[1]])
            return [prim_linea,sec_linea,ter_linea,cuar_linea]
        if self.f_angle == pi/3 and self.s_angle == 5*pi/3:
            prim_linea = Linea(self.f_angle, self.s_angle, [self.x_s[0], self.x_s[0]+self.dist_x()],[self.y_s[0], self.y_s[0]+self.dist_y()])
            sec_linea = Linea(self.f_angle + pi/3, self.f_angle - pi/3, [prim_linea.x_s[1], prim_linea.x_s[1]+self.dist()*math.cos(self.f_angle)],[prim_linea.y_s[1], prim_linea.y_s[1]+self.dist()*math.sin(self.f_angle)])
            ter_linea = Linea(self.s_angle+pi/3,self.s_angle -pi/3,[sec_linea.x_s[1],sec_linea.x_s[1] + self.dist()*math.cos(self.s_angle)],[sec_linea.y_s[1], sec_linea.y_s[1]+ self.dist()*math.sin(self.s_angle)])
            cuar_linea = Linea(self.f_angle, self.s_angle, [ter_linea.x_s[1], self.x_s[1]], [ter_linea.y_s[1],self.y_s[1]])
            return [prim_linea,sec_linea,ter_linea,cuar_linea]
        prim_linea = Linea(self.f_angle, self.s_angle, [self.x_s[0], self.x_s[0]+self.dist_x()],[self.y_s[0], self.y_s[0]+self.dist_y()])
        sec_linea = Linea(self.f_angle + pi/3, self.f_angle - pi/3, [prim_linea.x_s[1], prim_linea.x_s[1]+self.dist()*math.cos(self.f_angle)],[prim_linea.y_s[1], prim_linea.y_s[1]+self.dist()*math.sin(self.f_angle)])
        ter_linea = Linea(self.s_angle - pi/3, self.s_angle + pi/3,[sec_linea.x_s[1],sec_linea.x_s[1] + self.dist()*math.cos(self.s_angle)],[sec_linea.y_s[1], sec_linea.y_s[1]+ self.dist()*math.sin(self.s_angle)])
        cuar_linea = Linea(self.f_angle, self.s_angle, [ter_linea.x_s[1], self.x_s[1]], [ter_linea.y_s[1],self.y_s[1]])
        return [prim_linea,sec_linea,ter_linea,cuar_linea]
                                                                                         
```

```python
primLinea = Linea(5*pi/3, pi/3, [1,2], [1,1])
secLinea = Linea(2*pi/3, 0, [1,1.5], [1, 1+math.sqrt(0.75)])
terLinea = Linea(0,4*pi/3, [1.5,2],[1+math.sqrt(0.75), 1])

to_graph = []
lineas = [primLinea, secLinea, terLinea]
for item in lineas:
        for linea in item.dividir():
            to_graph.append(linea)
```

```python
depth=5;
conjunto_lineas=[lineas]
conjunto_iteracion =[]
for i in range(depth): ##depth
    for linea in conjunto_lineas[-1]: ##
        for elem in linea.dividir():
            conjunto_iteracion.append(elem)
    conjunto_lineas.append(conjunto_iteracion)
    conjunto_iteracion = []
```

```python
fig,ax = pl.subplots(figsize=(10,10))
for item in conjunto_lineas[5]:
    ax.plot(item.x_s, item.y_s)
    ax.set_ylim(0,3)
    ax.set_xlim(0,3)
```

## Propiedades

### Area

En la primera iteracion, pensando en solo un segmento, tendremos un triangulo equilatero de base $ \frac{1}{3} $. Mediante el Teorema de Pitagoras, deducimos que el area de este triangulo es:

### $$A_1 = \frac{1}{4}*(\sqrt{3}*a^2)$$

El tama√±o de a, que es la base, decrece $\frac{1}{3}$ por cada iteracion, por lo que, dada la iteracion $ i $ el area del triangulo sera

### $$A_i = \frac{1}{4}*(\sqrt{3}*(\frac{a}{3^i})^2)$$

En cada iteracion, se agregan dos nuevos triangulos que no forman parte del triangulo original, por lo que para obtener el area total, debemos multiplicar esta area por dos y sumar las areas anteriores
