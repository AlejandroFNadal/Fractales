## Triangulo de Pascal

Se forma comenzando por 1, luego con una fila de 1,1, y luego, vas sumando los pares de arriba para conformar una piramide. Primero, voy a generar los valores con Python

```python
from termcolor import colored
import numpy as np
import math
```

```python
iteraciones = 33
filas =[[1]]
temp_col = []
for i in range(1,iteraciones):
    for j in range(i):
        if(j==0):
            temp_col.append(filas[i-1][j])
        elif(j== i-1):
            temp_col.append(filas[i-1][j-1])
        else:
            temp_col.append(filas[i-1][j-1]+filas[i-1][j])
    filas.append(temp_col)
    temp_col = []
filas.pop(0)
```

```python
pad = math.ceil((iteraciones- len(filas[0]))/2)
np_arrays = np.array(np.pad(filas[0],(pad,pad+1),'constant',constant_values=(0,0)))
for item in filas:
    if(len(item) % 2 == 0):
        acum =0
    else:
        acum =1
    pad = math.ceil((iteraciones- len(item))/2)
    temp_np_arr=np.pad(item,(pad,pad+acum),'constant',constant_values=(0,0))
    print(temp_np_arr)
    print(temp_np_arr.shape)
    np_arrays=np.vstack((np_arrays,temp_np_arr))
print(np_arrays)
```

```python
for i in range(np_arrays.shape[0]):
    
    for j in range(np_arrays.shape[1]):
        if(np_arrays[i][j] !=0):
            if(np_arrays[i][j] % 2 == 0):
                print(colored(1,'white','on_blue'),end=' ')
            else:
                print(colored(0,'white','on_red'),end=' ')
        else:
            print(' ',end=' ')
    print('')
```

Esto ocurre porque, la suma de dos numeros pares da siempre un numero par, la suma de dos impares da par, y la suma de un par con un impar da impar. Como a los costados tenemos impares, esos generan regularmente un impar, que a la vez es capaz de generar otros impares en las filas sucesivas. Solo dejaran de generar impares cuando se choquen con otro impar que viene desde el otro lado en diagonal. Por ende, se terminan formando los triangulos.
